# SET OPERATIONS

```python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
```

```python
set1.union(set2)  # Union – combines all unique elements
set1 | set2
```

```python
set1.intersection(set2)  ## Intersection – common elements
set1 & set2
```

```python
set1-set2  ## Difference – in set1 but not in set2
set1.difference(set2)
```

```python
set1 + set2  #No addition
```

```python
# Symmetric Difference – in either set, but not both
print(set1 ^ set2)                     # {1, 2, 5, 6}
print(set1.symmetric_difference(set2))# same result
```

```python
set1.issubset(set2)
```

```python
set1.issuperset(set2)
```

```python
# 🌗 Think of it like this:

# Saying A is a subset of B means
# 👉 all elements of A are found inside B.

# Saying B is a superset of A means
# 👉 B contains all elements of A.

# So they’re literally two ways of describing the same fact — like saying:

# “Earth orbits the Sun” ☀️
# vs
# “The Sun is orbited by Earth.” 🌍

# Same relationship, flipped perspective.
# Subset = from A’s point of view.
# Superset = from B’s point of view.
```

```python
sub={1,3}
```

```python
sub.issubset(set1)
```

```python
set1.issuperset(sub)
```
