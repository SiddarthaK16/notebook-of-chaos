```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import root_mean_squared_error
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
df=pd.read_csv("Training_Set.csv")
```

```python
df_features= df.drop("Ending_Stocks" , axis=1)
df_labels=df["Ending_Stocks"].copy()
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(df_features)
```

```python
X=pd.DataFrame(
    X ,
    columns = df_features.columns,
    index=df_features.index
)
```

# LINEAR REGRESSOR

```python
linear=LinearRegression()
linear.fit(X,df_labels)
linear.predict(X)
linear_rmse=cross_val_score(
    linear,
    X,
    df_labels,
    scoring="neg_root_mean_squared_error",
    cv=10
)
print(-linear_rmse.mean())
```

```python
forest=RandomForestRegressor()
forest.fit(X,df_labels)
forest.predict(X)
forest_rmse=cross_val_score(
    forest,
    X,
    df_labels,
    scoring="neg_root_mean_squared_error",
    cv=10
)
print(-forest_rmse.mean())
```

```python
tree=DecisionTreeRegressor()
tree.fit(X,df_labels)
tree.predict(X)
tree_rmse=cross_val_score(
    tree ,
    X,
    df_labels,
    scoring="neg_root_mean_squared_error",
    cv=10
)
print(-tree_rmse.mean())
```

```python
neighbor=KNeighborsRegressor()
neighbor.fit(X,df_labels)
neighbor.predict(X)
neighbor_rmse=cross_val_score(
    neighbor ,
    X,
    df_labels,
    scoring="neg_root_mean_squared_error",
    cv=10
)
print(-neighbor_rmse.mean())
```

```python
grad=GradientBoostingRegressor()
grad.fit(X,df_labels)
grad.predict(X)
grad_rmse=cross_val_score(
    grad ,
    X,
    df_labels,
    scoring="neg_root_mean_squared_error",
    cv=10
)
print(-grad_rmse.mean())
```
