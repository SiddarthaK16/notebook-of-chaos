```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
data=pd.read_csv("breastCancer-train.csv")
data
```

```python
print(data.head())
print(data.tail())
```

```python
x=data.iloc[ : ,2:11 ]
x
```

```python
y=data.iloc[ : ,1]
y
```

```python
model=RandomForestClassifier()
```

```python
model.fit(x,y) #only based on all mean values.. feels acurate enough
```

```python
prediction=model.predict([[19.89,20.26,130.5,1214,0.1037,0.131,0.1411,0.09431,0.1802]])
if(prediction=="M"):
    print("Maligant , dangererous and tumuor may spread. be aware and critical care is needed")
elif(prediction=="B"):
    print("Benign , curable but be conscious and improve lifestyle")
```

```python
data_test=pd.read_excel("breastcancer-test.xlsx")
data_test
```

```python
x_test=data_test.iloc[ : ,2:11 ]
x_test
```

```python
y_test=data_test.iloc[ : ,1]
y_test
```

```python
predictions = model.predict(x_test.values)
predictions
```

```python
actual=y_test.values
```

```python
count=0
for i in range(0,len(actual)):
    if predictions[i]==actual[i]:
        count+=1
print((count/len(actual))*100)
```
