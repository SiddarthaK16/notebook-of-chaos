```python
from sklearn.datasets import make_classification
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(
    
    n_samples=1000,      
    # Total number of data points to generate
    
    n_features=2,        
    # Total number of input features (columns)
    # Since we want to visualize with a scatter plot, we keep only 2 features
    
    n_informative=2,     
    # Number of features that actually contain useful signal for predicting the class
    # Here both features help determine the class
    
    n_redundant=0,       
    # Redundant features are linear combinations of informative ones
    # Setting this to 0 avoids weird correlated diagonal patterns
    
    n_clusters_per_class=1,  
    # Number of clusters (groups) for each class
    # 1 → each class forms one main blob
    
    class_sep=2,         
    # Controls how far apart the classes are
    # Higher value → easier separation between classes
    
    weights=[0.9, 0.1],  
    # Controls class imbalance
    # 90% of samples will be class 0
    # 10% of samples will be class 1
    
    random_state=42      
    # Fixes randomness so the dataset is reproducible
    # Anyone running this code gets the same data
)
```

```python
df1=pd.DataFrame(X,columns= ["f1" , "f2"] )
df2=pd.DataFrame(y,columns=["target"])
df=pd.concat([df1,df2] , axis=1)
```

```python
plt.scatter(df["f1"] , df["f2"] , c=df["target"])
```
