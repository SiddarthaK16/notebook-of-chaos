```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

# ONE HOT ENCODING

```python
df=pd.DataFrame({
    "colour" : ["red","blue" , "black" , "green" , "yellow" , "orange"]
})
```

```python
df
```

```python
OHE_encoder=OneHotEncoder()
```

```python
OHE_encoded=OHE_encoder.fit_transform(df[["colour"]]).toarray()
```

```python
OHE_encoded
```

```python
OHE_encoded_df=pd.DataFrame(OHE_encoded , columns=OHE_encoder.get_feature_names_out())    #Returns the column names created by the encoder
OHE_encoded_df
```

# LABEL ENCODING

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE_encoder=LabelEncoder()
```

```python
LE_encoded=LE_encoder.fit_transform(df[["colour"]])
```

```python
LE_encoded
```

# ORDINAL ENCODING

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ranked=pd.DataFrame({
    "size" : ["small","medium" , "small" , "medium" , "large"]
})
```

```python
OE_encoder=OrdinalEncoder(categories=[["small","medium" , "large"]])
```

```python
OE_encoded=OE_encoder.fit_transform(ranked[["size"]])
```

```python
OE_encoded
```

# TARGET GUIDED ORDINAL ENCODING

```python
df = pd.DataFrame({
    'city' : ['New York', 'London', 'Paris', 'Tokyo', 'New York', 'Paris'],
    'price': [200, 150, 300, 250, 180, 320]})
```

```python
df
```

```python
mean_price=df.groupby("city")["price"].mean().to_dict()
mean_price
```

```python
df["city_encoded"] = df["city"].map(mean_price)
```

```python
df
```
