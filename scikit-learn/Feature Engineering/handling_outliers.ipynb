```python
import numpy as np
```

```python
virat_scores=[59,31,7,67,22,62,1,73,70,51,62,43,54,12,43,190,200,450]
minimum , Q1 , median , Q3 , maximum = np.quantile(virat_scores,[0.00,0.25,0.50,0.75,1.00])
```

```python
minimum , Q1 , median , Q3 , maximum
```

```python
IQR=Q3-Q1
lower= Q1 - (1.5 * (IQR))
upper = Q1 + (1.5*(IQR))
```

```python
lower,upper
```

```python
import seaborn as sns
```

```python
sns.boxplot(virat_scores)  # That dots are the outliers
```
