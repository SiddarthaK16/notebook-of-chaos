```python
a=requests.get("https://books.toscrape.com/")
```

```python
for i in range(1,51):
    a=requests.get(f"https://books.toscrape.com/catalogue/page-{i}.html")
    with open (f"books_scraped/page{i}.html" , "w" , encoding="utf8") as f:
        f.write(a.text)
    print(f"Page {i} printed successfully")
        
```
