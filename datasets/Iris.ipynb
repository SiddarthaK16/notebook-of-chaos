```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
iris = sns.load_dataset("iris")
```

```python
iris
```

```python
iris.info()
```

```python
iris.describe()
```

```python
iris.groupby("species")[['sepal_length' , 'sepal_width' , 'petal_length' , 'petal_width']].mean()
```

```python
colors=["red" , "orange" , "green"]
sns.set_theme(style="darkgrid")
sns.barplot(x="species" , y="sepal_length" , data=iris , palette=colors , ci=None)
plt.xlabel("Species")
plt.ylabel("Sepal length")
plt.title("Sepal lengths of different species")
plt.tight_layout()
plt.legend()
plt.grid(True)
```

```python
colors=["red" , "orange" , "green"]
sns.set_theme(style="darkgrid")
sns.barplot(x="species" , y="sepal_width" , data=iris , palette=colors , ci=None)
plt.xlabel("Species")
plt.ylabel("Sepal width")
plt.title("Sepal widths of different species")
plt.tight_layout()
plt.legend()
plt.grid(True)
```

```python
colors=["red" , "orange" , "green"]
sns.set_theme(style="darkgrid")
sns.barplot(x="species" , y="petal_length" , data=iris , palette=colors , ci=None)
plt.xlabel("Species")
plt.ylabel("Petal length")
plt.title("Petal lengths of different species")
plt.tight_layout()
plt.legend()
plt.grid(True)
```

```python
colors=["red" , "orange" , "green"]
sns.set_theme(style="darkgrid")
sns.barplot(x="species" , y="petal_width" , data=iris , palette=colors , ci=None)
plt.xlabel("Species")
plt.ylabel("Petal widht")
plt.title("Petal width of different species")
plt.tight_layout()
plt.legend()
plt.grid(True)
```

```python
sns.scatterplot(
    data=iris,
    x="petal_length",
    y="petal_width",
    hue="species",
    palette=colors,
    s=80,           # size of dots
    alpha=0.8       # slight transparency
)

plt.title("Petal Length vs Petal Width (Iris Dataset)")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.legend(title="Species")
plt.tight_layout()
plt.show()
```
