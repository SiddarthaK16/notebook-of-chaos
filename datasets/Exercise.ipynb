```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
exercise=sns.load_dataset("exercise")
exercise
```

```python
exercise.info()
```

```python
exercise.describe()
```

```python
sns.set_style("whitegrid")
sns.barplot(
    x="kind", y="pulse", data=exercise , hue="diet",
    palette="viridis", edgecolor="black", ci=None
)
plt.title("Pulse Variation Across Exercise Types", fontsize=14, weight="bold")
plt.xlabel("Type of Exercise")
plt.ylabel("Average Pulse (bpm)") 
plt.tight_layout()
```

```python
sns.set_style("whitegrid")
sns.barplot(
    x="time", y="pulse", data=exercise, hue="kind" , 
    palette="viridis", edgecolor="black", ci=None
)
plt.title("Pulse Variation Across Exercise duration", fontsize=14, weight="bold")
plt.xlabel("Duration of Exercise")
plt.ylabel("Average Pulse (bpm)")
plt.tight_layout()
```

```python
sns.set_style("whitegrid")
sns.barplot(
    x="time", y="pulse", data=exercise, hue="diet" , 
    palette="viridis", edgecolor="black", ci=None
)
plt.title("Pulse Variation Across Exercise duration", fontsize=14, weight="bold")
plt.xlabel("Duration of Exercise")
plt.ylabel("Average Pulse (bpm)")
plt.tight_layout()
```

```python
sns.barplot(x="diet", y="pulse", hue="time", data=exercise,
            palette="tab10", edgecolor="black", ci=None)
plt.title("Pulse Variation Across Diets and Durations", fontsize=12, weight="bold")
plt.xlabel("Diet Type")
plt.ylabel("Average Pulse (bpm)")
plt.legend(title="Exercise Duration")
```
