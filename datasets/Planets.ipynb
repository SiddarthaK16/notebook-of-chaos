```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
planets=sns.load_dataset("planets")
planets
```

```python
planets.info()
```

```python
planets.describe()
```

```python
planets1=planets.dropna()
```

```python
planets1
```

```python
planets1[planets1['method'] == "Radial Velocity"]
```

```python

sns.set_style(style="darkgrid")
sns.scatterplot(x="mass" , y="orbital_period", data=planets1 , alpha=0.7 )
plt.xlabel("MASS OF PLANET")
plt.ylabel("ORBITAL PERIOD OF THE PLANET")
plt.xscale("log")  #takes scattered large values to log size
plt.yscale("log")

plt.title("VARIATION OF ORBITAL PERIOD WITH MASS OF PLANET")

plt.grid(True)
plt.tight_layout()
```

### Most exoplanet catalogs (including seaborn’s planets dataset) measure mass in Jupiter masses, not Earth masses.

So:

* 10⁰ = 1 Jupiter mass

* 10¹ = 10 Jupiter masses

* 10⁻¹ = 0.1 Jupiter masses

* 10⁻² = 0.01 Jupiter masses

* Now put that in Earth units:

* 1
 Jupiter mass
 =
 318
* Earth masses
  1 Jupiter mass=318 Earth masses

* So:

 0.01 Jupiter masses = 3.18 Earth masses

```python

sns.set_style(style="darkgrid")
sns.scatterplot(x="distance" , y="orbital_period", data=planets1 , alpha=0.7 )
plt.xlabel("DISTANCE OF PLANET")
plt.ylabel("ORBITAL PERIOD OF THE PLANET")
plt.xscale("log")  #takes scattered large values to log size
plt.yscale("log")

plt.title("VARIATION OF ORBITAL PERIOD WITH DISTANCE OF PLANET")

plt.grid(True)
plt.tight_layout()
```

```python

sns.set_style(style="darkgrid")
sns.scatterplot(x="mass" , y="distance", data=planets1 , alpha=0.7 )
plt.xlabel("MASS OF PLANET")
plt.ylabel("DISTANCE OF THE PLANET FROM EARTH")
plt.xscale("log")  #takes scattered large values to log size
plt.yscale("log")

plt.title("DO HEAVIER PLANETS LIVE AWAY FROM EARTH? :)")

plt.grid(True)
plt.tight_layout()
```
