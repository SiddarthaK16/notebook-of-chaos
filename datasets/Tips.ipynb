```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
tips.info()
```

```python
sns.set_style(style="darkgrid")
sns.scatterplot(x="total_bill" , y="tip" , hue="sex" , data=tips)
plt.xlabel("Total bill")
plt.ylabel("Tip")
plt.title("TOTAL BILL vs TIP LEFT")
plt.grid(True)
plt.legend()
```

```python
sns.set_style(style="darkgrid")
sns.barplot(x="size" , y="tip" , hue="sex" , data=tips , ci=None)
plt.xlabel("Number of people in a table")
plt.ylabel("Tip")
plt.title("Number of people vs Tip left")
plt.grid(True)
plt.legend()
```

```python
tips[tips['size'] == 5].groupby('sex')['tip'].describe()
```

```python
sns.set_style("whitegrid")
sns.barplot(
    x="day", y="tip", hue="sex", data=tips,
    palette="rocket", edgecolor="black", alpha=0.9, ci=None
)
plt.title("Day vs Tip Left ðŸ’°")
plt.xlabel("Day of the Week")
plt.ylabel("Average Tip ($)")
plt.legend(title="Sex")
```

```python
sns.set_style("whitegrid")
sns.barplot(
    x="smoker", y="tip", hue="sex", data=tips,
    palette="rocket", edgecolor="black", alpha=0.9, ci=None
)
plt.title("Does smoking affects tipping ðŸ’°")
plt.xlabel("Type")
plt.ylabel("Average Tip ($)")
plt.legend(title="Sex")
```

```python
sns.set_style("darkgrid")
sns.barplot(
    x="time", y="tip", hue="sex", data=tips,
    palette="magma", edgecolor="black", alpha=0.9, ci=None
)
plt.title("Which time implies more tip")
plt.xlabel("Type of  mean")
plt.ylabel("Average Tip ($)")
plt.legend(title="Sex")
plt.tight_layout()
```

```python
sns.set_style("darkgrid")
sns.barplot(
    x="smoker", y="total_bill", hue="sex", data=tips,
    palette="magma", edgecolor="black", alpha=0.9, ci=None
)
plt.title("Do smokers eat more?")
plt.xlabel("Smoker or not")
plt.ylabel("Average Bill")
plt.legend(title="Sex")
plt.tight_layout()
```

```python
sns.set_style("darkgrid")

order = ["Thur", "Fri", "Sat", "Sun"]
sns.barplot(
    x="day", y="size", hue="smoker", data=tips,
    order=order, palette="flare", edgecolor="black", alpha=0.9, ci=None
)                                            

plt.title("When does smokers meetup happen?")
plt.xlabel("Day")
plt.ylabel("Number of people")
plt.legend(title="Smoker or not")
plt.tight_layout()
```
