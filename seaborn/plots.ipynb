```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
tips.info()
```

# LINEPLOTS

```python
sns.set_theme(style="dark")
sns.lineplot(x="total_bill" , y="tip" , data=tips)
plt.title("VARIATION OF TIPS WITH TOTAL BILL STUDY")
plt.tight_layout()
plt.grid(True)
```

# Scatter Plot

```python
sns.scatterplot(x="total_bill" , y="tip" , data=tips , hue="time") #hue = color code different groups in your data automatically.
plt.title("VARIATION OF TIPS WITH TOTAL BILL STUDY")
plt.tight_layout()
plt.grid(True)
```

# BARPLOT

```python
sns.barplot(x="day", y="total_bill", data=tips , hue="sex",ci=None) # ci removes that black error line
plt.title('Average Bill per Day')
plt.tight_layout()
```

# BOXPLOT

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title('Boxplot of Total Bill per Day')  # those dots show data way out 25-75th percentile
plt.tight_layout()
```

# HEATMAP

```python
flights = sns.load_dataset('flights')
pivot_table = flights.pivot(index="month",columns= "year",values= "passengers")

sns.heatmap(pivot_table, annot=True, fmt="d", cmap="YlGnBu")
#annot=True â†’ writes the actual numbers inside the boxes.
#fmt="d" â†’ formats them as integers (no decimals).
#cmap="YlGnBu" â†’ the aesthetic blue-green-yellow gradient.
plt.title('Heatmap of Passengers')
plt.tight_layout()
```
