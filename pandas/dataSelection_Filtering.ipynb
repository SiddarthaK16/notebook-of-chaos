# Selecting Rows & Columns

```python
df["Actor"]
```

```python
df["Actor"][1] # Single column (as Series)
```

```python
df[["Actor","BoxOffice in INR CRORE"]] # Multiple columns (as DataFrame)
```

```python
df.loc[1] ## First row (by label)
```

```python
df.iloc[1] # First row (by position)
```

```python
df.loc[3,"Film"]   # Value at row 3, column 'Film'
```

```python
print(df.iloc[3,1])  # Value at row 3, column at index 1
```

```python
df.loc[0:2 , ["Actor","Film","IMDb"] ]  # Rows 0 to 2, selected columns #2 inclusive
```

```python
df.iloc[0:2, 0:2]              # Rows and cols by index position  #2 exclusive(same as python)
```

```python
import time
```

```python
start=time.time()
df.loc[0,"Actor"]
end=time.time()

print(end-start)
print(df.loc[0,"Actor"])
```

```python
start=time.time()
df.at[0,"Actor"]    # Fast access to single element...
end=time.time()

print(end-start)
print(df.at[0,"Actor"])
```
