```python
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
plt.style.use("petroff10")
```

```python

players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs, labels=players, colors=colors)
plt.title("Career Runs of Indian Batsmen")
```

# wedgeprops

```python
plt.pie(
    runs,
    labels=players,
    colors=colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': '--' , 'fill':True}
)
plt.show()
```

# explode

```python
 explode = [0.2, 0, 0, 0]  # Only highlight Sachin's slice

plt.pie(runs, labels=players, explode=explode, colors=colors)
plt.title("Exploded Pie Example")
plt.show()
```

# Add Shadows for a 3D Feel

```python
plt.pie(
    runs,
    labels=players,
    explode=explode,
    colors=colors,
    shadow=True,  # adds a 3D-like shadow
    startangle=12, #rotates pie chart with that angle
    autopct="%1.1f%%"  #shows percentage
)
plt.show()
```
