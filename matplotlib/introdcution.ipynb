```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4,5,6] , [4,6,10,12,14,20])
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]


plt.xlabel("Year")
plt.ylabel("Runs scored")
plt.title("Runs scored by Sachin Tendulkar over the years") 
plt.plot(years,runs)
```

# Multiple lines in a plot

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, label="Virat Kohli")  # label is better than using legend
plt.plot(years, sehwag, label="Virender Sehwag")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
# plt.legend()
```

# FORMAT STRINGS

```python
plt.plot(years, kohli, 'o--r', label="Kohli")  # red circles with dashed lines
plt.plot(years, sehwag, '8:g', label="Sehwag")  # green triangles dotted
# plt.legend()  #fmt = '[marker][line][color]' is the format
```

# Color and Line Style Arguments

```python
plt.plot(years, kohli, color='orange', linestyle='--', label="Kohli")
plt.plot(years, sehwag, color='green', linestyle='-.', label="Sehwag")
plt.plot(years, runs, color='blue', label="Tendulkar")
```

```python
plt.plot(years, kohli, color="green" , linestyle="-" , linewidth=3, label="Kohli")
plt.plot(years, sehwag, color="red" , linestyle="-" , linewidth=2, label="Sehwag")

plt.grid(True)   #shows grids in the plot
plt.tight_layout()  #plt.tight_layout() swoops in and says,“Yo, move over, make room for everyone!”


```

```python
plt.style.available  # shows styles available
```

```python
plt.style.use('fivethirtyeight') #to use selected style

plt.plot(years, kohli, color="green" , linestyle="-" , linewidth=3, label="Kohli")
plt.plot(years, sehwag, color="red" , linestyle="-" , linewidth=2, label="Sehwag")
```

# XKCD Comic Style

```python
with plt.xkcd():
    plt.plot(years, kohli, label="Kohli")
    plt.plot(years, sehwag, label="Sehwag")
    plt.title("Epic Battle of the Batsmen")
    plt.grid(True)
```
