```python
import matplotlib.pyplot as plt
```

```python

# Data
x = [1, 2, 3, 4, 5]
y1 = [i * 2 for i in x]
y2 = [i ** 2 for i in x]

# Create a figure with 1 row and 2 columns
plt.subplot(1, 2, 1)  # (rows, cols, plot_no)
plt.plot(x, y1)
plt.title('Double of x')

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title('Square of x')

plt.tight_layout()
```

```python
plt.plot(x,y1 , label="Double of x" , color="red" , linestyle="-" , linewidth=2)
plt.plot(x,y2 , label="Square of x" , color="blue" , linestyle="--" , linewidth=3)

plt.xlabel("Number")
plt.ylabel("Value")
plt.title("Double vs Square of a number")   # I FEEL THIS IS BETTER
plt.tight_layout()
plt.legend()
plt.grid(True)
```

#  2×2 Grid of Subplots

```python
# Data
x = [1, 2, 3, 4, 5]
y1 = [i * 2 for i in x]
y2 = [i ** 2 for i in x]
y3 = [i ** 0.5 for i in x]
y4 = [10 - i for i in x]
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.xlabel("Number")
plt.ylabel("Double of number")
plt.title("Double of a number")
plt.tight_layout()
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.xlabel("Number")
plt.ylabel("Square of number")
plt.title("Square of a number")
plt.tight_layout()
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.xlabel("Number")
plt.ylabel("Square root of number")
plt.title("Square root of a number")
plt.tight_layout()
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.xlabel("Number")
plt.ylabel(" 10-Number")
plt.title("10-Number")
plt.tight_layout()
plt.grid(True)
```

# Using plt.subplots() for Clean Code

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
 
axs[0].plot(x, y1)
axs[0].set_title('x * 2')
 
axs[1].plot(x, y2)
axs[1].set_title('x squared')
 
fig.suptitle('Simple Comparison Plots', fontsize=14)
fig.tight_layout()
fig.subplots_adjust(top=0.85)  # So title doesn't overlap
fig.savefig('my_plots.png')    # Save as image
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
ys = [y1, y2, y3, y4]
titles = ['x * 2', 'x squared', 'sqrt(x)', '10 - x']

for i in range(2):
    for j in range(2):
        idx = i * 2 + j
        axs[i, j].plot(x, ys[idx])
        axs[i, j].set_title(titles[idx])

plt.tight_layout()
plt.show()
```
