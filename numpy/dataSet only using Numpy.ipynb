```python
import numpy as np

# Seed for reproducibility
np.random.seed(101)

# Big array: 100 rows, 6 columns
# Columns: Age, BMI, Blood Pressure, Glucose, Insulin, Outcome (0/1)
data = np.random.randint(20, 80, size=(100, 1))   # Age 20-79
bmi = np.random.randint(15, 40, size=(100, 1))    # BMI 15-39
bp = np.random.randint(80, 160, size=(100, 1))    # Blood Pressure 80-159
glucose = np.random.randint(70, 200, size=(100, 1)) # Glucose 70-199
insulin = np.random.randint(15, 276, size=(100, 1)) # Insulin 15-275
outcome = np.random.randint(0, 2, size=(100, 1))    # 0 or 1

# Stack them horizontally
data = np.hstack((data, bmi, bp, glucose, insulin, outcome))

print("Shape of data:", data.shape)
print("First 5 rows:\n", data[:5])
```

```python
print(data)
```

## Compute the mean, median, and standard deviation of BMI and Glucose.

```python
bmi=data[:,1]
glucose=data[:,3]

mean1=np.mean(bmi)
mean2=np.mean(glucose)

sd1=np.std(bmi)
sd2=np.std(glucose)

print(mean1)
print(mean2)
print(sd1)
print(sd2)
```

## How many diabetics (Outcome = 1) have Blood Pressure > 120?

```python
bp=data[:,2]

high= bp>120

print(np.size(bp[high]))
```

## Find the 25th, 50th, and 75th percentiles of Age

```python
age=data[:,0]

print(np.percentile(age,25))
print(np.percentile(age,50))
print(np.percentile(age,75))
```

## Which row has the maximum Glucose value? What are all the feature values of that person?

```python
print(np.argmax(glucose))

print(data[np.argmax(glucose)])
```

## Normalize the columns Age, BMI, and Blood Pressure to a [0,1] scale using Min-Max normalization.

```python
# Columns: Age=0, BMI=1, BP=2
cols_to_norm = data[:, :3]  # first 3 columns
min_vals = cols_to_norm.min(axis=0)
max_vals = cols_to_norm.max(axis=0)

normalized = (cols_to_norm - min_vals) / (max_vals - min_vals)
print(normalized)    # normalise=(data-min) / (max-min)
```
